```python
import pandas as pd
df1 = pd.read_csv('SWWQDiscovery_clean.csv')
df2 = pd.read_csv('SWActiveSites_fishing.csv')
```

```python
df2
```

```python
# merge the datasets
df1 = df1.drop('CurrentSite', axis=1)
df3 = df2[['STATION', 'STNAME', 'LATITUDE', 'LONGITUDE', 'FISHING', 'FISH']]
df4 = pd.merge(df3, df1, how='left', on='STATION')
```

```python
df4
```

```python
# drop and reset index
df4 = df4[df4['LATITUDE'] != 0]
df4 = df4.reset_index()
df4 = df4.drop('index', 1)
df4
```

```python
# split the STNAME
split_data = df4["STNAME"].str.split(" @ ")
data = list(split_data)
names = ["RIVER", "LOCATION"]
new_df = pd.DataFrame(data, columns=names)
new_df.fillna(value=pd.np.nan, inplace=True)
print(new_df)
```

```python
# update the data
df4['STNAME'] = new_df['RIVER']
df4['LOCATION'] = new_df['LOCATION']
df4['FISHING'] = df4['FISHING'].astype(int)
df4['FISH'] = df4['FISH'].str.capitalize()
df4
```

```python
df4.to_csv('final.csv', index=False)
```
