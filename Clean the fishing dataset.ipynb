```python
# get data
import wget
url = "http://data.water.vic.gov.au/wini/state/SWActiveSites.csv"
file = wget.download(url)
```

```python
# clean the data
import pandas as pd
df = pd.read_csv('SWActiveSites.csv')
df = df[1:920]
df.to_csv('SWActiveSites_clean.csv', index=False)
```

```python
# read the data
df1 = pd.read_csv('SWActiveSites_clean.csv')
df2 = pd.read_csv('tmpgme16Yrecreational-fishing-spots.csv', encoding = "ISO-8859-1")
```

```python
# create list for STNAME
stname = list(df1['STNAME'])
stname
```

```python
# match the locations
import re
from itertools import repeat
location = []
fish = []
for i in range(len(df2)):
    r = re.compile(".*"+df2['Location'][i].upper()+"*.")
    newlist = list(filter(r.match, stname))
    if newlist != []:
        location.extend(newlist)
        fish.extend(repeat(df2['Fish'][i],len(newlist)))
```

```python
# update the data
df1['FISHING'] = 0
for i in range(len(df1)):
    if df1['STNAME'][i] in location:
        df1['FISHING'][i] = 1
```

```python
# match fish with location
fishing = {'STNAME': location,
          'FISH': fish
          }

df3 = pd.DataFrame(fishing, columns = ['STNAME', 'FISH'])

df4 = pd.merge(df1, df3, how='left', on='STNAME')
```
