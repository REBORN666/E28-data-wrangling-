```python
import wget
import pandas as pd
import pymysql
import csv
```

```python
# get data
url = "http://data.water.vic.gov.au/wini/state/SWWQDiscovery.csv"
file = wget.download(url)
```

```python
# clean data
df = pd.read_csv('SWWQDiscovery.csv')
df = df[1:552]
df.to_csv('SWWQDiscovery_clean.csv', index=False)
```

```python
# create connection
con = pymysql.connect(host="127.0.0.1",
                      user="root",
                      passwd="REBORN")
cur = con.cursor()

cur.execute('create database testdb')
con.select_db('testdb')
cur.execute('create table testtable (STATION int,pH int)')
```

```python
# import data
with open('SWWQDiscovery_clean.csv') as csvfile:
    sp = csv.DictReader(csvfile)
    for row in sp:
        sql = 'insert into testtable (STATION,pH) values (%d,%d)'%(int(row['STATION']),int(row['pH']))
        print(sql)
        cur.execute(sql)
con.commit()
cur.close()
```
